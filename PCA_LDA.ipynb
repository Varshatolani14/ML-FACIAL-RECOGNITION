# **Face Recognition Using PCA & LDA**

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os
from PIL import Image
```

# Generate the Data Matrix and the Label vector

```python
paths=['datasets/s'+str(i) for i in range(1,41)]
cnt=0
Data=np.zeros((400,10304))
labels=np.zeros((400,1))
for i in range(40):
    labels[i*10:(i+1)*10]=i+1  
for path in paths:
    files = os.listdir(path)
    for file in files:
        img = Image.open(path + "/" + file)
        np_img = np.array(img)
        np_img = np_img.flatten()
        Data[cnt]=np_img
        cnt+=1
print(labels)        
            
```

```python
image_height = 112
image_width = 92
image_array = np.reshape(Data[0], (image_height, image_width))

# Plot the image
plt.imshow(image_array, cmap='gray')  # Use cmap='gray' for grayscale images
plt.axis('off')  # Turn off axis
plt.show()
```
